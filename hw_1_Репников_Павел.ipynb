```python
result = df1
result.rename(columns={'ID': 'ID_CLIENT'}, inplace=True)
result = pd.merge(result, df3, on='ID_CLIENT', how='outer')
result = pd.merge(result, df4, on='ID_CLIENT', how='outer')
result = pd.merge(result, df7, on='ID_CLIENT', how='outer')
result = pd.merge(result, df8, on='ID_CLIENT', how='outer')
```

```python
df_merged = pd.merge(df5, df2, on='ID_LOAN', how='outer')
df_merged = df_merged.groupby('ID_CLIENT').agg(
    LOAN_NUM_TOTAL=('ID_LOAN', 'count'),
    LOAN_NUM_CLOSED=('CLOSED_FL', 'sum')
).reset_index()
df_merged
```

```python
result = pd.merge(result, df_merged, on='ID_CLIENT', how='outer')
df = result
df = df.drop_duplicates()
df
```

```python
df = df.drop(['REG_ADDRESS_PROVINCE', 'FACT_ADDRESS_PROVINCE', 'POSTAL_ADDRESS_PROVINCE', 'AGREEMENT_RK'], axis=1)
```
